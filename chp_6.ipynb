**6.1** Thresholding based segmentation

````python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
# Read the image in grayscale
img = cv2.imread('/content/butterfly.jpeg')
#Apply Simple thresholding
rets, thresholds = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
# Apply Otsu's Thresholding
ret, threshold = cv2.threshold(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the result
cv2_imshow(thresholds)
cv2_imshow(threshold)

# Apply segmentation
contours, hierarchy = cv2.findContours( threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the image
segmented_image = cv2.drawContours(img, contours, -1, (255, 255, 0), 2)

# Display result
cv2_imshow(segmented_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

````

````python


# Apply Mean Adaptive Thresholding
threshold_mean = cv2.adaptiveThreshold(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY), 255, cv2.ADAPTIVE_THRESH_MEAN_C,  cv2.THRESH_BINARY, 11, 2)

# Apply Gaussian Adaptive Thresholding
threshold_gaussian = cv2.adaptiveThreshold(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Display the results
cv2_imshow(threshold_mean)
cv2_imshow(threshold_gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()

````

6.2 K means Clustering

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (can replace with any image from your dataset)
image = cv2.imread('/content/butterfly.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Reshape the image to a 2D array of pixels
pixels = image.reshape((-1, 3))

# Convert to float32 for kmeans
pixels = np.float32(pixels)

# Define criteria and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
k = 3  # Number of clusters (segments)
_, labels, centers = cv2.kmeans(pixels, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Convert centers back to 8 bit values
centers = np.uint8(centers)

# Map each pixel to the center of the cluster
segmented_image = centers[labels.flatten()]
segmented_image = segmented_image.reshape(image.shape)

# Display the original and segmented images
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(segmented_image)
plt.title("Segmented Image (K-means)")
plt.show()

````

````python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def split_and_merge(image, threshold, min_size=10):
    """
    Performs region splitting and merging on an image.

    Args:
        image: The input grayscale image.
        threshold: The homogeneity threshold for splitting.
        min_size: The minimum size of a region (prevents splitting too far).

    Returns:
        A list of segmented regions (as NumPy arrays).
    """

    h, w = image.shape
    regions = []

    def split(image, depth=0, max_depth=10):
        """Recursively splits regions."""
        # Check if the region is empty or too small or reached maximum recursion depth
        if image.size == 0 or image.shape[0] < min_size or image.shape[1] < min_size or depth >= max_depth:
            return [image]

        if np.max(image) - np.min(image) < threshold:  # If homogeneous, no split
            return [image]
        else:
            mid_h, mid_w = h // 2, w // 2
            regions = []
            regions += split(image[:mid_h, :mid_w], depth + 1)  # Top-left
            regions += split(image[:mid_h, mid_w:], depth + 1)  # Top-right
            regions += split(image[mid_h:, :mid_w], depth + 1)  # Bottom-left
            regions += split(image[mid_h:, mid_w:], depth + 1)  # Bottom-right
            return regions

    regions = split(image)
    return regions

# Load an example image
image = cv2.imread('/content/download.jfif', cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (128, 128))

# Step 2: Apply the splitting and merging
threshold = 50
regions = split_and_merge(image, threshold)

# Step 3: Display the resulting regions
for i, region in enumerate(regions):
    print(f"Region {i + 1}:")
    cv2_imshow( region)  # Use cv2.imshow

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load an example image (replace 'example_image.jpg' with your image file)
image = cv2.imread('/content/download.jfif',cv2.IMREAD_COLOR)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 1: Apply Canny Edge Detection
edges = cv2.Canny(gray, 10, 10)  # Adjust thresholds (50 and 150) as needed

# Step 2: Threshold the edges to create a binary image
_, binary = cv2.threshold(edges, 20, 70, cv2.THRESH_BINARY)

# Step 3: Find markers (initial regions based on binary edges)
markers = np.zeros_like(gray)
markers[binary == 255] = 1  # Mark edge regions as 1
markers = markers.astype(np.int32)

# Step 4: Perform Watershed
cv2.watershed(image, markers)

# Step 5: Mark boundaries with a distinct color
image[markers == -1] = [0, 0, 255]  # Mark watershed boundaries in red

# Display the result
cv2_imshow(image)

cv2.waitKey(0)
cv2.destroyAllWindows()
````
